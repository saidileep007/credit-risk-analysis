## Imputing Null Values

Imputing null values is the process of trying to fill the null values with mean/median/mode/any other ways of collecting data to fill the null/missing values, thus helping us get additional information from the missing rows and columns. Inputing null values is possible only when there are lower number of null values compared to the ob=verall dataset. If there are too many null/missing values, we drop the rows/columns. One point to keep in mind is that dropping a row/column should be the last resort as this will cause certain import features to loose information that might be dependant on these columns/rows, or they might be important columns/rows by themselves.

## Encoding

Machine learning models cannot undestand non numeric languages, hence we need to convert our non numeric columns into numeric columns before passing the dataset to our model, such that the model can unederstand the dataset. There are 3 types of encoding, these are label encoding/One hot encoding/frequency encoding. We use Label encoding when we want to give importance to the word ie, encoded based on alphabetic order. We use One hot encoding/dummy encoding when we have a binary categorical column and lastly we use frequency encoding when we have too many categories in the column and we dont want to create additional columns for all the categories as it is done is dummy encoding

## Inference from the Base Model

From all the above models, we can see that there is a severe drawdown in the train prediction and test prediction scores. This is mainly because of the imbalance in the target variable. This is evident from the classification report as the 0 class is predicted very well in some models, where as the 1 class is not predicted as expected. This is because there is a huge imbalance of 40% in the dataset, where the 0 class has 74% of the datapoints and the 1 class only has 26% of the datapoints. This can be rectified using the SMOTE function in python.
